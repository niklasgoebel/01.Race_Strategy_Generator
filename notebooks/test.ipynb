```python
import numpy as np
from src.loaders.gpx_loader import load_gpx_to_df

gpx_path = "/Users/niklasgoebel/Desktop/Projects/Race Strategy Generator/data/chianti course.gpx"  # <-- replace
df = load_gpx_to_df(gpx_path)

raw = df["elev_raw"].to_numpy(dtype=float)
smooth = df["elev_smooth"].to_numpy(dtype=float)
x_km = df["cum_distance"].to_numpy(dtype=float) / 1000.0

mask = np.isfinite(raw) & (raw <= 0)

print("count raw <= 0:", int(mask.sum()))

if mask.sum() > 0:
    idxs = np.where(mask)[0]

    runs = []
    start = idxs[0]
    prev = idxs[0]
    for i in idxs[1:]:
        if i == prev + 1:
            prev = i
        else:
            runs.append((start, prev))
            start = prev = i
    runs.append((start, prev))

    runs = sorted(runs, key=lambda t: (t[1]-t[0]+1), reverse=True)

    print("Top bad runs:")
    for a, b in runs[:5]:
        seg = raw[a:b+1]
        print(
            f"indices {a}-{b} | len={b-a+1} | "
            f"km {x_km[a]:.3f}-{x_km[b]:.3f} | "
            f"min={np.nanmin(seg):.1f}, max={np.nanmax(seg):.1f}"
        )
else:
    print("No raw <= 0 values found")
```
